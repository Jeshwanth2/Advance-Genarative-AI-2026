##2nd Problem

```python
marks=[45, 78, 90, 33, 60]
pass_count=0
fail_count=0
for score in marks:
    if score>=50:
        pass_count+=1
    else:
        fail_count+=1
print("Total Pass Students",pass_count)
print("Total Fail Students",fail_count)
```

# 4. Message Length Analyzer


```python
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
   print(f"Message: {message}, Length: {len(message)}")
   if len(message)>10:
    print("message is long")
```

#5 Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error=0
for log in logs:
    if log=="ERROR":
        error+=1
print("Total Error Messages",error)
```
